```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cols = ['fLength','fWidth','fSize','fConc','fConc1','fAsym','fM3Long','fM3Trans','fAlpha','fDist','class'] #this will basically add headings or lets say variable name in our data
df= pd.read_csv('magic04.data', names=cols)
df.head()
```

```python
df["class"] = (df["class"] == "g").astype(int)
```

```python
df.head()
```

```python
for label in cols[:-1]:
    plt.hist(df[df["class"] == 1 ] [label] , color = 'blue',label='gamma',alpha= 0.7 ,density = True)
    plt.hist(df[df["class"] == 0 ] [label] , color = 'red',label='gamma',alpha= 0.7 ,density = True)
    plt.title(label)
    plt.ylabel('Probablity')
    plt.xlabel(label)
    plt.legend()
    plt.show()
```
